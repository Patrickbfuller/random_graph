```python
plt.figure(figsize=(22,18))
for _ in range(1000):
    a = np.random.randint(0,1000,5)
    b = np.random.randint(0,1000,5)
    hexlist = [str(np.random.randint(0,10)) for i in range(6)]
    color = '#'+''.join(hexlist)
    marker = np.random.choice(list(markers.keys()))
    alpha = np.random.random()**2.5
    size = np.random.exponential()**5.3
    plt.scatter(a, b, c=color, marker=marker,
                s=size, alpha=alpha**0.7)
    plt.axis('off')
```
